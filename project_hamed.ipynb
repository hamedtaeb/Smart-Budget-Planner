### Cleaning

```python
## Clean Categorical Columns
df['Marital_Status'] = df['Marital_Status'].str.strip()
df['Education'] = df['Education'].str.strip()
```

```python
# Fixing Wired Martial_Status Entries
martial_map = {
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single',
}

df['Marital_Status_norm'] = df['Marital_Status'].replace(martial_map)


# Normalize 'Education' Entries
edu_map = {
    '2n Cycle': 'Graduate',
    'Graduation': 'Graduate',
    'Master': 'Post-Graduate',
    'PhD': 'Post-Graduate',
    'Basic': 'Basic'
}

df['Education_norm'] = df['Education'].replace(edu_map)
```

```python
# Parse Data
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce', dayfirst=True)
```

```python
# Impute Missing Income 
df['Income']= pd.to_numeric(df['Income'], errors='coerce')
group_median = df.groupby(['Education_norm', 'Marital_Status_norm'])['Income'].transform('median')
df['Income'] = df['Income'].fillna(group_median)
df['Income'] = df['Income'].fillna(df['Income'].median())

print("Remaining nulls in income:", df['Income'].isnull().sum())
df[['Education', 'Education_norm', 'Marital_Status', 'Marital_Status_norm']].head(10)
```

### Feature Engineering

```python
# Reference date - Today for our age & tenure calculations
REF_DATE = pd.Timestamp.today().normalize()

# Calculate Customer Age
df['Age'] = REF_DATE.year - df['Year_Birth']

# Children QTY
df['Children'] = df['Kidhome'] + df['Teenhome']

# Customer Tenure  : Measures loyalty or relationship age
df['Customer_For_Days'] = (REF_DATE - df['Dt_Customer']).dt.days
df['Customer_For_years'] = (df['Customer_For_Days'] / 365.25).round(1)


# Total Spending ( Sum of all Mnt* columns)
mnt_cols = [col for col in df.columns if col.startswith('Mnt')]
df['TotalMnt'] = df[mnt_cols].sum(axis=1)

# Total Purchase 
purchase_cols = [col for col in df.columns if col.startswith('Num') and col.endswith('Purchases')]
df['TotalPurchases'] = df[purchase_cols].sum(axis=1)
```

```python

# Total Purchase 
purchase_cols = [col for col in df.columns if col.startswith('Num') and col.endswith('Purchases')]
df['TotalPurchases'] = df[purchase_cols].sum(axis=1)

# Average Spend Per Purchase 
df["AvgMntPerPurchase"] = np.where(df['TotalPurchases'] > 0,
                                   df['TotalMnt'] / df['TotalPurchases'], 
                                   np.nan)

# Replacing rows with 0 transactions to 1 as not possible to generate a spend with no txs
df["TotalPurchases"] = df["TotalPurchases"].replace(
    {
        0 : 1
    }
)
```

```python
# CHANNEL SHARES
for channel in ['Web', 'Catalog', 'Store']:
    col = f'Num{channel}Purchases'
    if col in df.columns:
        df[f'{channel}PurchaseShare'] = np.where(df['TotalPurchases'] > 0,
                                                 df[col] / df['TotalPurchases'],
                                                 np.nan)
        


# Deal Purchase Rate
df['DealPurchaseRate'] = np.where(df['TotalPurchases'] > 0,
                                  df['NumDealsPurchases'] / df['TotalPurchases'],
                                  np.nan)


# Campaign Response Rate
cmp_cols = [col for col in df.columns if col.startswith('AcceptedCmp')]
df['CampaignsAccepted'] = df[cmp_cols].sum(axis=1)
df['Responded'] = (df['Response'] == 1).astype(int)
```

```python
# Creating flags

def flag_by_percentile(series, low_q=0.33, high_q=0.67):
    low, high = series.quantile([low_q, high_q])
    return pd.cut(series, bins=[-float('inf'), low, high, float('inf')], labels=['Low', 'Medium', 'High'])


df['Income_level'] = flag_by_percentile(df['Income'])
df['Spending_level'] = flag_by_percentile(df['TotalMnt'])
df['Deal_level'] = flag_by_percentile(df['DealPurchaseRate'])
df['Online_level'] = flag_by_percentile(df['WebPurchaseShare'])
df['Response_level'] = pd.cut(         #### Beacuse most of the values are zero we use cut instead of quantile(I had an error priror to it)
    df['CampaignsAccepted'],
    bins=[-np.inf, 0, 1, np.inf],
    labels=['Low', 'Mid', 'High'],
    include_lowest=True
)


# Sanity Check
df[['Income_level', 'Spending_level', 'Deal_level', 'Online_level', 'Response_level']].head(10)
```

```python
# Applying Personalyti Logis

def classify_personality(row):
    # Saver: deal-seeking, modes income/spend
    if row['Deal_level'] == 'High' and row['Income_level'] == 'High':
        return 'Saver'  
    
    # Splurger: high spenders, High Income
    elif row['Income_level'] == 'High' and row['Spending_level'] == 'High' and row['Deal_level'] != 'High':
        return 'Splurger'
    
    #Impulse Buyer: high online share, high deal rate
    elif row['Online_level'] == 'High' and row['Deal_level'] == 'Low':
        return 'Impulse Buyer'
    
    # Planner: balanced or engaged with campaigns
    elif row['Response_level'] == 'High' or (row['Deal_level'] == 'Medium' and row['Online_level'] == 'Medium'):
        return 'Planner'
    
    else: 
        return 'Planner' # Default to Planner
    

df['Personality_Type'] = df.apply(classify_personality, axis=1)
df['Personality_Type'].value_counts()
```

### K-Mean Clustering

```python
# Select Core features for clustering

cluster_features = [
    'Income', 'TotalMnt',         # For Spending Capacity
    'TotalPurchases', 'AvgMntPerPurchase',   # For Activity Level(Purchase Frequency)
    'DealPurchaseRate',                      # For Deal Sensitivity (Impulse vs Planner)
    'WebPurchaseShare', 'CatalogPurchaseShare', 'StorePurchaseShare'  # For Channel Preference
]

X= df[cluster_features].copy()

# Since I got some erorr for Nan Values 
prep = Pipeline([
    ("imputer", SimpleImputer(strategy='median')),
    ("scaler", StandardScaler())
    ])

X_scaled = prep.fit_transform(X)
```

```python
# Testing different K values to find optimal K
inertia = []
K =range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)


# Plotting the Elbow
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters K')
plt.ylabel('Intertia')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
# Fit K-Means with k=5
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)


# Check cluster size
df['Cluster'].value_counts().sort_index()
```

```python
# Compare Cluster Means
cluster_summary = df.groupby('Cluster')[cluster_features].mean().round(2)
display(cluster_summary)
```

###  K-Means Clustering: Finding Natural Spending Groups

#### Goal
The goal of clustering was to **discover natural groups of customers** based on their *spending behavior* — without using predefined personality labels.  
This helps validate whether our rule-based categories (Planner, Splurger, Saver, etc.) actually exist in the data.

#### Step 1: Selecting Behavioral Features
We focused on numerical features that describe *how* customers spend:

| Feature | Description |
|----------|--------------|
| **Income** | Customer’s financial capacity to spend |
| **TotalMnt** | Total amount spent across all product categories |
| **TotalPurchases** | Frequency of purchases |
| **AvgMntPerPurchase** | Average amount per purchase — reflects buying power |
| **DealPurchaseRate** | Tendency to buy discounted products |
| **WebPurchaseShare** | Preference for online shopping |
| **CatalogPurchaseShare** | Preference for catalog purchases |
| **StorePurchaseShare** | Preference for in-store purchases |

These capture three key behavioral dimensions:
1. **Capacity** (Income)  
2. **Engagement** (TotalMnt, TotalPurchases)  
3. **Style** (Deal usage & purchase channel preference)



#### Step 2: Scaling the Data
All features were standardized using `StandardScaler()` to ensure they are comparable.  
This is essential because **K-Means uses Euclidean distance**, and unscaled data (e.g., Income vs. DealRate) would bias the results.



#### Step 3: Determining Optimal K (Number of Clusters)
We used the **Elbow Method**, plotting inertia (within-cluster variance) for K = 2–9.  
The curve flattened around **K = 5**, indicating that five clusters provide the best trade-off between compactness and simplicity.



#### Step 4: Running K-Means (K = 5)
We applied K-Means with 5 clusters.  
Each customer was assigned to the nearest centroid — representing the “average” spending pattern of that group.

### PCA Visualization of Customer Clusters

#### What is PCA?

**Principal Component Analysis (PCA)** is a dimensionality reduction technique used to **project high-dimensional data into a lower-dimensional space** (usually 2D or 3D) while keeping as much information (variance) as possible.

Since our spending dataset has multiple correlated features (Income, TotalMnt, DealRate, etc.), PCA helps:
- Summarize those features into a few composite axes (called *principal components*).
- Visualize the **structure and separation of clusters** that would otherwise be hidden in higher dimensions.

Each axis (PCA1, PCA2) is a linear combination of the original features and represents the directions where the data varies the most.



#### What the PCA Plot Shows

The scatter plots display each customer as a point in a **2D PCA space**, colored by their assigned cluster (0–4).  
Clusters that are **tightly grouped** represent customers with similar spending habits, while **separated clusters** indicate distinct behavioral patterns.

From the plots:

- **Cluster 2** (Splurgers) is clearly more concentrated — high-spend, high-income individuals.  
- **Cluster 1** (Planners) and **Cluster 3** (Savers) occupy different PCA regions, showing distinct shopping behaviors.  
- **Cluster 0** appears somewhat scattered, representing lower-income, low-spend groups.  
- **Cluster 4** is an **isolated outlier**, confirming its extreme spending pattern (a single customer with one large purchase).



#### Interpretation

The PCA visualization validates that:
- The **K-Means model successfully formed meaningful clusters** in spending behavior.
- The **Splurger and Planner** groups are relatively well-defined in feature space.
- **Savers** form a broader, more diverse segment — typical in real-world consumer data.
- Outliers (like Cluster 4) are easily identifiable when visualized in reduced dimensions.

Together, K-Means + PCA offer both **quantitative segmentation** and **visual confirmation** that our customer groups are behaviorally distinct.
